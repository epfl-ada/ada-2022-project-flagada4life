# Pre-analysis :

## Age :

```python
agedf = df.copy()
#Remove nan value and actor age below 0
agedf.dropna(subset = ['Actor age at movie release', 'Actor gender'],inplace=True)
agedf = agedf[agedf['Actor age at movie release'].apply(lambda x: x > 0)]
```

```python
agedf.boxplot(by='Actor gender', column=['Actor age at movie release'], grid=False, showmeans=True,)
```

```python
agedf.dropna(subset = ['Movie release date'],inplace=True)
agedf.sort_values(by='Movie release date',ascending=True,inplace=True) #The pd.to_datetime will sort automatically the values
agedf['Movie release year'] = pd.to_datetime(agedf['Movie release date']).dt.year
```

```python
men = agedf[agedf['Actor gender'].apply(lambda x : x == 'M')]
women = agedf[agedf['Actor gender'].apply(lambda x : x == 'F')]
```

```python
def bootstrap_CI(data, nbr_draws):
    means = np.zeros(nbr_draws)
    data = np.array(data)

    for n in range(nbr_draws):
        indices = np.random.randint(0, len(data), len(data))
        data_tmp = data[indices] 
        means[n] = np.nanmean(data_tmp)

    return [np.nanpercentile(means, 2.5),np.nanpercentile(means, 97.5)]
```

```python
stats_by_year_men = men.groupby(men['Movie release year']).apply(lambda x: pd.Series({
        'average_age': x['Actor age at movie release'].mean(),
        'lower_err_age': bootstrap_CI(x['Actor age at movie release'], 1000)[0],
        'upper_err_age': bootstrap_CI(x['Actor age at movie release'], 1000)[1]
    }))

stats_by_year_women = women.groupby(women['Movie release year']).apply(lambda x: pd.Series({
        'average_age': x['Actor age at movie release'].mean(),
        'lower_err_age': bootstrap_CI(x['Actor age at movie release'], 1000)[0],
        'upper_err_age': bootstrap_CI(x['Actor age at movie release'], 1000)[1]
    }))
```

```python
plt.fill_between(stats_by_year_men.index, stats_by_year_men.lower_err_age,
                 stats_by_year_men.upper_err_age, alpha = 0.5, color = 'blue')
plt.plot(stats_by_year_men.average_age, color = 'black')

plt.fill_between(stats_by_year_women.index, stats_by_year_women.lower_err_age,
                 stats_by_year_women.upper_err_age, alpha = 0.5, color = 'red')
plt.plot(stats_by_year_women.average_age, color = 'grey')

plt.xlabel('Year')
plt.ylabel('Age at movie release')
```

## Number of Men/Women

```python
nbr_men_women = df.copy()
nbr_men_women.dropna(subset = ['Movie release date'],inplace=True)
nbr_men_women.sort_values(by='Movie release date',ascending=True,inplace=True) #The pd.to_datetime will sort automatically the values
nbr_men_women['Movie release year'] = pd.to_datetime(nbr_men_women['Movie release date'], errors='coerce').dt.year
nbr_men_women.query('`Movie release year` >= 1895', inplace=True)
nbr_men_women.dropna(subset = ['Movie release year'],inplace=True) #remove invalide release year, before 1895 (ex : 1010)
nbr_men_women['Movie release year'] = nbr_men_women['Movie release year'].astype(int)
```

```python
nbr_men_women['Actor gender'].replace('', np.nan, inplace=True)
nbr_men_women.dropna(subset = ['Actor gender'], inplace=True)

men_women_by_film=nbr_men_women.groupby(nbr_men_women['Wikipedia movie ID']).apply(lambda x : pd.Series({
    'number of women' : x['Actor gender'].apply(lambda x: x == 'F').sum(),
    'number of men' : x['Actor gender'].apply(lambda x: x == 'M').sum(),
    'Movie release year' : x['Movie release year'].unique()[0]
    }))
```

```python
ax = men_women_by_film[['number of women','number of men']].boxplot(grid=False, showmeans=True)
ax.set_ylim(0, 30)
```

```python
men_women_by_film
```

```python
stats_by_film_men = men_women_by_film.groupby(men_women_by_film['Movie release year']).apply(lambda x: pd.Series({
        'average_men': x['number of men'].mean(),
        'lower_err_men': bootstrap_CI(x['number of men'], 1000)[0],
        'upper_err_men': bootstrap_CI(x['number of men'], 1000)[1]
    }))

stats_by_year_women = men_women_by_film.groupby(men_women_by_film['Movie release year']).apply(lambda x: pd.Series({
        'average_women': x['number of women'].mean(),
        'lower_err_women': bootstrap_CI(x['number of women'], 1000)[0],
        'upper_err_women': bootstrap_CI(x['number of women'], 1000)[1]
    }))
```

```python
plt.fill_between(stats_by_film_men.index, stats_by_film_men.lower_err_men,
                 stats_by_film_men.upper_err_men, alpha = 0.5, color = 'blue')
plt.plot(stats_by_film_men.average_men, color = 'black')

plt.fill_between(stats_by_year_women.index, stats_by_year_women.lower_err_women,
                 stats_by_year_women.upper_err_women, alpha = 0.5, color = 'red')
plt.plot(stats_by_year_women.average_women, color = 'grey')

plt.xlabel('Year')
plt.ylabel('Number of men/women by film')
```
